```python
##小明上菜啦 IV 虛擬碼
def LumberPuzzle(lumberlst):
    if sum(lumberlst) % 4 != 0:
        return False
    
    length = sum(lumberlst) / 4
    remainlst = [int(length), int(length), int(length), int(length)]
    for edge in remainlst:
        uselst = []
        index = 0
        left_length = edge
        while True:
            if left_length == 0:
                break
            try:
                if left_length >= lumberlst[index]:
                    left_length -= lumberlst[index]
                    uselst.append(lumberlst[index])
            except: # index 走到底了，都沒有找到合適的木材拼完
                if uselst == []:
                    return False

            index += 1
        
        for i in uselst:
            lumberlst.remove(i)


lumberlst = [int(x) for x in input().split(",")]
lumberlst = sorted(lumberlst)[::-1]
print(LumberPuzzle(lumberlst))
```

```python
#小明上菜啦 II 助教demo
namedict = {}

while True:
    try:
        name = input()
        name, alph =name, tuple(sorted(name.lower()))
        if not namedict.get(alph):
            namedict[alph] = name
            print("Create " + name)
        elif name == namedict.get(alph):
            print(name + " already exists")
        else:
            old_name = namedict[alph]
            print(name + " is " + old_name + " , " + old_name + " already exists")
            
    except EOFError:
        break
            
```

```python
##小明上菜啦 IV 歐陽碼 Accepted(雖然沒有搞懂...)
def LumberPuzzle(lumberlst):
    k = len(lumberlst)
    if sum(lumberlst) % 4 != 0:
        return False
    
    length = sum(lumberlst) / 4
    remainlst = [0, 0, 0, 0]
    def dfs(index):
        if index == k:
            return remainlst[0] == remainlst[1] == remainlst[2] == length
        for i in range(4):
            if remainlst[i] + lumberlst[index] <= length:
                remainlst[i] += lumberlst[index]
                if dfs(index + 1):
                    return True
                remainlst[i] -= lumberlst[index]
        return False
    return dfs(0)


lumberlst = [int(x) for x in input().split(",")]
lumberlst = sorted(lumberlst)[::-1]
print(LumberPuzzle(lumberlst))
```
